```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Libraries imported successfully!")
```

```python
file_path = r"C:\Users\parth\Downloads\airline_passenger_satisfaction.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
mean_flight_distance = df['Flight Distance'].mean()

print("The average Flight Distance is:", mean_flight_distance)
```

```python
median_flight_distance = df['Flight Distance'].median()

print("The median Flight Distance is:", median_flight_distance)
```

```python
flight_distance_variance = df['Flight Distance'].var()

print("The variance of Flight Distance is:", flight_distance_variance)
```

```python
flight_distance_std = df['Flight Distance'].std()

print("The standard deviation for Flight Distance is:", flight_distance_std)
```

```python
flight_distance_sem = df['Flight Distance'].sem()

print("The standard error of the mean for flight_distance is:", flight_distance_sem)
```

```python
# This creates a simple histogram instantly
flight_distance_hist = df['Flight Distance'].hist()
print(flight_distance_hist)
```

```python
# This prints a single number telling you the skew
df['Flight Distance'].skew()
```

```python
print("Mean  :", mean_flight_distance)
print("Median:", median_flight_distance)

if mean_flight_distance > median_flight_distance:
    print("\nDistribution is Right-Skewed (Positively Skewed)")
    
elif mean_flight_distance < median_flight_distance:
    print("\nDistribution is Left-Skewed (Negatively Skewed)")
    
else:
    print("\nDistribution is Symmetric (Normal Distribution)")
```

```python
# 1. Choose your column
column = df['Flight Distance']

# 2. Find the 25th percentile (Q1) and 75th percentile (Q3)
q1 = column.quantile(0.25)
q3 = column.quantile(0.75)

# 3. Calculate the Interquartile Range (IQR)
iqr = q3 - q1

# 4. Define the upper and lower boundaries for "normal" data
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

# 5. Filter the dataframe to isolate the outliers
outliers = df[(column < lower_bound) | (column > upper_bound)]

# 6. Print the results clearly
print(f"--- Outlier Detection for Flight Distance ---")
print(f"Normal range is between: {lower_bound:.2f} and {upper_bound:.2f}")
print(f"Total number of outlier rows found: {len(outliers)}")
```
